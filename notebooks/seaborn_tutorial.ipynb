```python
#Seaborn tutorial
#https://seaborn.pydata.org/tutorial/relational.html
#load packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
#oal dataset
tips = sns.load_dataset("tips")

#scatter is the default plot type in the function replot in seaborn
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
#add a third dimension to the plot for categorical data
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
#adust the marker image
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips);
```

```python
#if you set a numeric variable to the hue, it will automatically change to a gradiant 
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
```

```python
#to customize a color palette

sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips);
```

```python
#another variable can be added, size, and customized
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips);
```

```python
#lineplots are created by setting kind to "line"
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.figure.autofmt_xdate()
```

```python
# you can plot the mean and 95% CI for the line plot as well, if you have many observations for one x
fmri = sns.load_dataset("fmri")
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```python
#disabling the bootstraped CI is also possible 
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri);
```

```python
#plotting the standard deviation of the mean instead of the CI
sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);
```

```python
#lineplots can also be split by categorical data
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);
```

```python
#other kinds of plots 
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
#kernal density estmator
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
#displot- "facetable", so you can quickly split the categorical data
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species")
```

```python
sns.pairplot(data=penguins, hue="species")
```
